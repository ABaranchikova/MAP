```python
from string import punctuation
punct = punctuation+'«»—…“”*№–'

def normalize(text):
    
    words = [word.strip(punct) for word in text.lower().split()]
    return words
```

```python
from tqdm import tqdm
```

```python
def spch (df):
    total_score = []
    with tqdm(total=len(df['text_norm'])) as pbar:
        for words in df['text_norm']:
            score = 0
            for word in words:
                check = Word(word)
                if check.correct is True:
                    pass
                else:
                    score+=1
            pbar.update(1)
            total_score.append(score)
    return total_score
            
            
```

```python
df_in["text_norm"] = df_in["text"].apply(normalize)
```

```python
df_out["text_norm"] = df_out["text"].apply(normalize)
```

```python
df_hold["text_norm"] = df_hold["text"].apply(normalize)
```
